1. Загрузка необходимых библиотек

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

2. Загрузка нечеткого изображения

```python
image = cv2.imread('brcde.jpg', 0)
```

3. Применение порогового значения к изображению с помощью cv2.threshold():

    В данном случае в качестве метода установления порога используется cv2.THRESH_BINARY + cv2.THRESH_OTSU. 
    Флаг cv2.THRESH_OTSU вычисляет оптимальное пороговое значение, используя метод Otsu.

```python
_, thresholded_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

4. Отображение исходного и порогового изображения с помощью matplotlib:

```python
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(thresholded_image, cmap='gray')
plt.title('Thresholded Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
